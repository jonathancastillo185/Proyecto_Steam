# Funciones utilizadas

```python
def extract(x): # ---> para desanidar los generos de games
    tags = []
    for x in x:
        tags.append(x)
    filtro = []
    for x in tags:
        for y in x.split(','):
            y = y.strip(" '[]'")
            filtro.append(y)
    return list(set(filtro))
```
