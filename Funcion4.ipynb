```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'data_set_limpio/games_preparado.csv.gz')
item = pd.read_csv(r'data_set_limpio/items_preparado.csv.gz')
opinion = pd.read_csv(r'data_set_limpio/reviews_preparado.csv.gz')
```

```python
def extraer_anio(fecha):
    if '-' in fecha:  # Si el formato es 'YYYY-MM-DD'
        return fecha.split('-')[0]
    else:  # Si el formato es 'MMM YYYY'
        partes = fecha.split(' ')
        if len(partes) == 2:
            return partes[1]
    return fecha
```

```python
prueba1 = pd.merge(opinion[['item_id','opinion','user','recommend']], df[['release_date','id','developer']], left_on='item_id', right_on='id')

prueba1['recommend'] = prueba1['recommend'].replace({True: 1, False: 0})

prueba1['release_date'] = prueba1['release_date'].apply(extraer_anio)

for index, row in prueba1.iterrows():
    if row[1] == 2 and row[3] == 1:
        prueba1.at[index, 'Positivo'] = 1
    else:
        prueba1.at[index, 'Positivo'] = 0

prueba1['Positivo'] = prueba1['Positivo'].astype(int)

prueba1
```

```python
anio = prueba1['release_date'].unique()
```

```python
anio
```

```python
end4 = {}
for x in anio:
    end4[x] = dict(prueba1.loc[(prueba1['release_date'] == x) & (prueba1['Positivo'] == 1)]['developer'].value_counts())
```

```python
# Supongamos que este es tu diccionario original

# Creamos un nuevo diccionario para almacenar los resultados
filtered_dict = {}

# Iteramos sobre cada clave y valor en el diccionario original
for key, value in end4.items():
    filtered_dict[key] = dict(list(value.items())[:3])

filtered_dict
```

```python
top1 = []
top2 = []
top3 = []

for x in anio:
    names = list(filtered_dict[x].keys())
    if len(names) >= 1:
        top1.append((names[0], filtered_dict[x][names[0]]))
    if len(names) >= 2:
        top2.append((names[1], filtered_dict[x][names[1]]))
    if len(names) >= 3:
        top3.append((names[2], filtered_dict[x][names[2]]))
```

```python
import pandas as pd

# AsegÃºrate de que todas las listas tengan la misma longitud
max_len = max(len(anio), len(top1), len(top2), len(top3))
anio = list(anio)
top1 = list(top1)
top2 = list(top2)
top3 = list(top3)

# Rellena las listas con None si es necesario
anio.extend([''] * (max_len - len(anio)))
top1.extend([''] * (max_len - len(top1)))
top2.extend([''] * (max_len - len(top2)))
top3.extend([''] * (max_len - len(top3)))

# Crea el DataFrame
impo2 = pd.DataFrame(data={'Anio': anio, 'Top 1': top1, 'Top 2': top2, 'Top 3': top3})
```

```python
impo2.to_csv('data_set_limpio\Max_developer_year.csv', index=False)
```
