```python
df = pd.read_csv(r'data_set_limpio/games_dummies.csv')
```

```python
def extract(x): # ---> para desanidar los generos de games
    tags = []
    for x in x:
        tags.append(x)
    filtro = []
    for x in tags:
        for y in x.split(','):
            y = y.strip(" '[]'")
            filtro.append(y)
    return list(set(filtro))
```

```python
def dumies_games(x): # ---> aca van los tags
    dicc = {}

    for x in list(x):
        dicc[x] = ''
        for y in df['tags']:
            if x in y:
                dicc[x] += '1'
            else:
                dicc[x] += '0'
    dicc1 = {}
    for x,e in dicc.items():
        dicc1[x] = [x for x in e]
        
    return pd.DataFrame(dicc1)
```

```python
def dumies_specs(x): # ---> aca van los tags
    dicc = {}

    for x in list(x):
        dicc[x] = ''
        for y in df['specs']:
            if x in y:
                dicc[x] += '1'
            else:
                dicc[x] += '0'
    dicc1 = {}
    for x,e in dicc.items():
        dicc1[x] = [x for x in e]
        
    return pd.DataFrame(dicc1)
```

```python
generos = extract(df['tags'])
```

```python
dumies = dumies_games(generos)
```

```python
result = pd.concat([df, dumies], axis=1)
```
